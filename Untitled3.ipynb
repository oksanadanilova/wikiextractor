```python
# Была задача построить гистограмму символов/биграм/слов в файле 20617.txt
# Файл содержит пачку юникодных строк, русские буквы, без мусора.
%matplotlib inline

from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import codecs

f = codecs.open('C:/Users/Oksana/JupiterFolder/Wiki/1-Original_script/wiki_00', encoding = 'utf-8')
s = f.read().split('\n')
f.close()

cv_letters = CountVectorizer(analyzer='char')

letters = cv_letters.fit_transform(s).toarray()
lett = zip([t[0] for t in sorted(cv_letters.vocabulary_.items(), key=lambda t: t[1])], np.array(letters.sum(axis=0)).squeeze())
```

```python
# let it be alphabetic
# просто быстрый код, не надо вчитываться:
labels, y = zip(*lett)
y = np.array(y)
x = np.arange(len(y))
plt.figure(figsize=(12, 9))
width = 0.3
rects1 = plt.bar(x, y, width, color='r')
plt.gca().set_xticks(x + width / 2)
plt.gca().set_xticklabels( labels )

plt.xlim([-0.25, max(x) + 0.5])
plt.show()
```
